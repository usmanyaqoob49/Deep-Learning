**Basic Functioning of Perceptron:**

perceptron functioning.jpg

Our Features for training a perceptron will be cgpa and resume_score.

And we will predict whether a person is placed or not.

```python
features= df[['cgpa', 'resume_score']]
labels= df['placed']
```

Importing Perceptron from Scikit learn Library:

```python
from sklearn.linear_model import Perceptron
```

```python
#initializing its object
p= Perceptron()
```

```python
#training perceptron on our data so that it can fit a line to differentiate classes
p.fit(features, labels)
```

```python
#weights that were used
p.coef_
```

Shows that W1 that is for cgpa has huge importance while predicting the Persons Placement.

```python
#bias value
p.intercept_
```

Lets plot how it drew the line:

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(features.values, labels.values, clf=p, legend=2)
```
