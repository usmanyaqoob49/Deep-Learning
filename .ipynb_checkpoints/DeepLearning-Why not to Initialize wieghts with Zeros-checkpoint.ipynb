```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from mlxtend.plotting import plot_decision_regions
```

# Creating Dataset:

```python
x, y= make_moons(n_samples= 100, 
                 noise= 0.1,
                 random_state= 42)

plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

# Training Deep Learning Model:

## When Hidden Layer Activation is ReLu:


\begin{array}{l}
a_{11} Activation Function \rightarrow \operatorname{Relu} \\
a_{11}=\max \left(0, z_{11}\right) \\
z_{11}=w_{11}^{\prime} x_{1}+w_{22}^{\prime} \times x_{2}+b_{11} \\
         w_{11}^{\prime}=0\ & w_{21}^{\prime}=0
\end{array}
\
\begin{array}{l}
a_{11}=0 \\
a_{12}=0
\end{array}


We will have to find the derivative of a11 and a12 in order to minimize loss using gradient descent and what will happen is that it will make that derivate term in Gradient Descent zero and so New Weight = Old Weight (and it will stay zero)



```python
model_relu = Sequential()

model_relu.add(Dense(10,activation='relu',input_dim=2))
model_relu.add(Dense(1,activation='sigmoid'))

model_relu.summary()
```

```python
#weights of the model that are initialized automatically by keras
weights= model_relu.get_weights()
weights
```

### Making all the Weights Zero:

```python
#weights values for layer 1 (2x10)
weights[0]= np.zeros(model_relu.get_weights()[0].shape)

#bias values for layer 1 (10, )
weights[1]= np.zeros(model_relu.get_weights()[1].shape)

#weights values for layer 2 (10x1)
weights[2]= np.zeros(model_relu.get_weights()[2].shape)

#bias values for layer 2 (1, )
weights[3]= np.zeros(model_relu.get_weights()[3].shape)
```

```python
weights
```

```python
#setting the new weights as the weights of the model
model_relu.set_weights(weights)
```

```python
#compiling
model_relu.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#training
history = model_relu.fit(x,y,epochs=100,validation_split=0.2)
```

### Plotting Decision Boundary:

```python
plot_decision_regions(x,y.astype('int'), clf=model_relu, legend=2)
plt.show()
```

**What a power model we have got with these weights**

### Weights after Training:

```python
model_relu.get_weights()
```

**You can see there is no change in weights so that means no learning or training was done**

## When Hidden Layer Activation is Tanh:

For tanh: 
\
\begin{array}{l}
=a_{11}=\frac{e^{z_{11}}-e^{-x_{11}}}{e^{211}+e^{-211}}=\frac{e^{0}-e^{0}}{e^{0}+e^{0}} \\
a_{11}=\frac{1-1}{1+1}=0 \\
a_{11}=0,a_{12}=0
\end{array}
\
And same thing that happen in relu will be repeated a these veights will stay zero No training.

```python
model_tanh = Sequential()
model_tanh.add(Dense(10,activation='tanh',input_dim=2))
model_tanh.add(Dense(1,activation='sigmoid'))


#stroing weights
weights= model_tanh.get_weights()

#changing weights to zero

#weights values for layer 1 (2x10)
weights[0]= np.zeros(model_tanh.get_weights()[0].shape)

#bias values for layer 1 (10, )
weights[1]= np.zeros(model_tanh.get_weights()[1].shape)

#weights values for layer 2 (10x1)
weights[2]= np.zeros(model_tanh.get_weights()[2].shape)

#bias values for layer 2 (1, )
weights[3]= np.zeros(model_tanh.get_weights()[3].shape)
```

```python
#setting models weights
model_tanh.set_weights(weights)
```

```python
model_tanh.get_weights()
```

```python
#compiling
model_tanh.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#training
history = model_tanh.fit(x,y,epochs=100,validation_split=0.2)
```

### Plotting Decision Boundary:

```python
plot_decision_regions(x,y.astype('int'), clf=model_tanh, legend=2)
plt.show()
```

**Same Poor Results, it is taking whole dataset as one category**

### Weights after Training:

```python
model_tanh.get_weights()
```

## When Hidden Layer Activation is Sigmoid:

```python
model_sigmoid = Sequential()
model_sigmoid.add(Dense(10,activation='sigmoid',input_dim=2))
model_sigmoid.add(Dense(1,activation='sigmoid'))


#stroing weights
weights= model_sigmoid.get_weights()

#changing weights to zero
#weights values for layer 1 (2x10)
weights[0]= np.zeros(model_sigmoid.get_weights()[0].shape)
#bias values for layer 1 (10, )
weights[1]= np.zeros(model_sigmoid.get_weights()[1].shape)
#weights values for layer 2 (10x1)
weights[2]= np.zeros(model_sigmoid.get_weights()[2].shape)
#bias values for layer 2 (1, )
weights[3]= np.zeros(model_sigmoid.get_weights()[3].shape)

#setting models weights
model_sigmoid.set_weights(weights)

#compiling
model_sigmoid.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#training
history = model_sigmoid.fit(x,y,epochs=100,validation_split=0.2)
```

### Plotting the Decision Boundary:

```python
plot_decision_regions(x,y.astype('int'), clf=model_sigmoid, legend=2)
plt.show()
```

```python
model_sigmoid.get_weights()
```

**You can clearly see that weights of input 1 for 10 nodes are same, and weights of input 2 for 10 nodes are same, so all these nodes will never the less act as single node.**
